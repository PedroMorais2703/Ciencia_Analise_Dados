# 3) Faça a avaliação do modelo com os dados de treinamento.
Traga a acurácia, recall e plote a matriz de confusão. Não se esqueça de avaliar com suas palavras o desempenho do modelo, interpretando as métricas.

Dica: Para calcularmos o recall em classificação multi classe precisamos usar o atributo macro:
recall = recall_score(y_train, y_pred_train, average='macro')

```python
y_train = y_train["Credit Score_High"]
y_test = y_test["Credit Score_High"]
```

```python
y_train
```

```python
y_test
```

```python
# Criação de uma instância do classificador Naive Bayes Gaussiano
naive_churn = GaussianNB()
# Treinamento do classificador usando os dados de treinamento
naive_churn.fit(X_train, y_train)
```

```python
# Fazer previsões para os dados de treino
y_pred_train = naive_churn.predict(X_train)
# Calcular a acurácia
accuracy = accuracy_score(y_train, y_pred_train)
print("Acurácia:", accuracy)
```

```python
# Matriz de confusão (use TESTE para análise real)
y_pred = naive_churn.predict(X_test)
conf_matrix = confusion_matrix(y_test, y_pred)


# Nomes das classes (ordem deve bater com os rótulos)
class_names = ['Score baixo ou medio', 'Score Alto']

# Plotando a matriz de confusão
fig = ff.create_annotated_heatmap(
    z=conf_matrix,
    x=class_names,
    y=class_names,
    colorscale='Blues',
    showscale=True
)

fig.update_layout(
    title='Matriz de Confusão',
    xaxis_title='Predicted labels',
    yaxis_title='True labels',
    font=dict(
        family='Arial',
        size=12,
        color='white'
    )
)
```

```python
conf_matrix = confusion_matrix(y_train, y_pred_train)

conf_matrix
```

```python
recall = recall_score(y_train, y_pred_train, average='macro')
print("Recall:", recall)
```

O modelo apresentou uma boa acuracia e recall mostrando que ele consegue idenditificar e clafissificar corretamente a maior parte dos registros da nossa base, já a matriz de confusão permite a melhor visualização dos dados.
